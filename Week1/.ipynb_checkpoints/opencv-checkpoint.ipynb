```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
path = "./opencv-logo.png"  

def img_opencv(path, window_name, key):
    image = cv2.imread(path)
    cv2.imshow(window_name, image)
    cv2.waitKey(key) 
    cv2.destroyAllWindows() 

def img_matplotlib(path, axis):
    image = mpimg.imread(path)
    plt.axis(axis)
    plt.imshow(image)
    plt.show()
    
def img_opencv2matplotlib(path, axis):
    image = cv2.imread(path)
    plt.axis(axis)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
img_opencv(path, 'logo', 0)
img_matplotlib(path, 'off')
img_opencv2matplotlib(path, 'off')
```
