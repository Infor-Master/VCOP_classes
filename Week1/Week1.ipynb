```python
#4)
#4.1)

x = np.arange(0, 4*np.pi, 0.1)   # start,stop,step
y1 = np.sin(x)
y2 = np.sin(x + np.pi)
y3 = np.sin(2 * x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()

#4.2)

x = np.arange(0, 10, 0.1)
y_lin = 4 * x
y_log = np.log(x + 2)
y_qud = np.power(x, 2)
plt.plot(x,y_lin)
plt.plot(x,y_log)
plt.plot(x,y_qud)
plt.show()
```
